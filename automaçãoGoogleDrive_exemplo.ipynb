```python
import pyautogui, time

pyautogui.alert("Por favor não mexa em nada no seu computador enquanto o codigo está rodando.")
pyautogui.PAUSE = 0.5
```

```python

pyautogui.position()
```

```python
#Abrir o google drive no meu computador 

#Alterna de janela
pyautogui.hotkey('alt', 'tab')
#Abre uma nova guia no navegador
pyautogui.hotkey('ctrl', 't')
time.sleep(0.5)
pyautogui.write("https://drive.google.com/drive/u/0/home")
pyautogui.press('enter')
time.sleep(5)
#Mover o mouse pra aba desejada 
pyautogui.moveTo(119,389)
#Selecionar o arquivo que quero baixar e clicar nele
pyautogui.moveTo(434,517)
pyautogui.click(button='right')
pyautogui.moveTo(618,599)

#Fazer o download do arquivo
pyautogui.click(button='left')
```
